```python
df_BM_orderbook['MidPrice'] = (df_BM_orderbook['askPrice']+df_BM_orderbook['bidPrice'])/2
df_Aax_orderbook['MidPrice'] = (df_Aax_orderbook['ask_1_price']+df_Aax_orderbook['bid_1_price'])/2

Bit_min = df_BM_orderbook.groupby(pd.Grouper(key='datetime', freq='1min')).min()['MidPrice'].values
Bit_mean = df_BM_orderbook.groupby(pd.Grouper(key='datetime', freq='1min')).mean()['MidPrice'].values
Bit_max = df_BM_orderbook.groupby(pd.Grouper(key='datetime', freq='1min')).max()['MidPrice'].values
Bit_datetime = df_BM_orderbook.groupby(pd.Grouper(key='datetime', freq='1min')).max().index

Aax_min = df_Aax_orderbook.groupby(pd.Grouper(key='datetime', freq='1min')).min()['MidPrice'].values
Aax_mean = df_Aax_orderbook.groupby(pd.Grouper(key='datetime', freq='1min')).mean()['MidPrice'].values
Aax_max = df_Aax_orderbook.groupby(pd.Grouper(key='datetime', freq='1min')).max()['MidPrice'].values
Aax_datetime = df_Aax_orderbook.groupby(pd.Grouper(key='datetime', freq='1min')).max().index
```

```python
df_Bt_bars = pd.DataFrame(columns=['datetime','Min Price','Mean Price','Max Price'])
df_Bt_bars['datetime'] = Bit_datetime
df_Bt_bars['Min Price'] = Bit_min
df_Bt_bars['Mean Price'] = Bit_mean
df_Bt_bars['Max Price'] = Bit_max

df_Aax_bars = pd.DataFrame(columns=['datetime','Min Price','Mean Price','Max Price'])
df_Aax_bars['datetime'] = Aax_datetime
df_Aax_bars['Min Price'] = Aax_min
df_Aax_bars['Mean Price'] = Aax_mean
df_Aax_bars['Max Price'] = Aax_max

df_Aax_bars = df_Aax_bars.dropna()
df_Bt_bars = df_Bt_bars.dropna()
df_Aax_bars.index = range(len(df_Aax_bars))
df_Bt_bars.index = range(len(df_Bt_bars))
```

```python
df_Bt_bars['D'] = df_Bt_bars['Max Price']-df_Bt_bars['Min Price']
df_Aax_bars['D'] = df_Aax_bars['Max Price']-df_Aax_bars['Min Price']
```

```python
df_Bt_bars['Price changes'] = df_Bt_bars['Mean Price'].diff()
mid_price = df_Bt_bars['Mean Price'].values
Price_changes_percent = [0]

for i in tqdm(range(1, len(df_Bt_bars))):
    a = ((mid_price[i]-mid_price[i-1])/mid_price[i-1])*100
    Price_changes_percent.append(a)
df_Bt_bars['Price changes percent'] = Price_changes_percent

df_Aax_bars['Price changes'] = df_Aax_bars['Mean Price'].diff()
mid_price = df_Aax_bars['Mean Price'].values
Price_changes_percent = [0]
for i in tqdm(range(1, len(df_Aax_bars))):
    a = ((mid_price[i]-mid_price[i-1])/mid_price[i-1])*100
    Price_changes_percent.append(a)
df_Aax_bars['Price changes percent'] = Price_changes_percent
```

```python
df_price_changes_Bt = df_Bt_bars[df_Bt_bars['Price changes percent']>0.05]
```

```python
df_PC_Aax = pd.DataFrame(columns=df_Aax_bars.columns)
for i in tqdm(range(len(df_price_changes_Bt))):
    a = df_Aax_bars[(df_Aax_bars['datetime']>=str(pd.to_datetime(df_price_changes_Bt['datetime'].values[i])-
                                                  timedelta(seconds=10)))&
                    (df_Aax_bars['datetime']<str(pd.to_datetime(df_price_changes_Bt['datetime'].values[i])+
                                                 timedelta(minutes=10)))]
    df_PC_Aax = pd.concat([df_PC_Aax,a])
df_PC_Aax = df_PC_Aax.drop_duplicates()
del a
```

```python
dict_PC = {}
for i in tqdm(range(len(df_price_changes_Bt))):
    a = df_Aax_bars[(df_Aax_bars['datetime']>=str(pd.to_datetime(df_price_changes_Bt['datetime'].values[i])-
                                                  timedelta(seconds=10)))&
                    (df_Aax_bars['datetime']<str(pd.to_datetime(df_price_changes_Bt['datetime'].values[i])+
                                                 timedelta(minutes=10)))]
    if len(a) != 0:
        dict_PC.update({df_price_changes_Bt['datetime'].iloc[i].strftime('%Y-%m-%d %H:%M:%S'): a})
del a
```
